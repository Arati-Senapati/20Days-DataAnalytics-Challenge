```python
import matplotlib.pyplot as plt

brands = ["Oneplus","Apple","Samsung","Nokia","Redmi"]
x = [22,35,30,3,10]
c=["red","violet","blue","magenta","orange"]
ex=[0,0,0,0,0.1]
plt.pie(x,labels = brands,colors=c,explode=ex,shadow=True,autopct="%.2f",startangle=180) #explode means highlight
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

data=pd.read_excel("expense3.xlsx")
df = pd.DataFrame(data)
print(df)
```

```python
plt.pie(df["Amount"])
plt.show()
```

```python
grouped_by = df.groupby("Payment Mode")["Amount"].sum()
print(grouped_by)
ex=[0,0.1,0]
plt.pie(grouped_by.values,labels= grouped_by.index,autopct ="%.2f",explode =ex)
plt.show()
```
