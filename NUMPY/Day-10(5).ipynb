```python
arr1 =[100,150,199,200,250,130]
arr2 =[10,50,30,40,30,10]

price = np.array(arr1)
quantity = np.array(arr2)

print(price,"\n",quantity)
```

```python
print(np.cumprod([price,quantity], axis = 0))
```

```python
c =np.cumprod([price,quantity], axis = 0)
print(c[1])
```
